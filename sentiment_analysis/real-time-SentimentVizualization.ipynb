# Real Time Vizualization Of Election Tweets

This code is the vizualization of the real time sentiment analysis created in the: real time sentiment analysis project.

In this project, we simply extract the real-time cummulative sum of the senttiments of both presidential candidates and plt them on a vizual graph.

To do this, we use the matplotlib animation library in conjunction with other libraries

```python
from itertools import count
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
%matplotlib notebook
```

```python
plt.style.use('fivethirtyeight')
```

Create a real-time animated graph that updates itself as the dataset updates. This graph plots the sentiment trends of the two candidates

```python
frame_length = 10000   # only use a dataset of 10000 rows at a time
fig = plt.figure(figsize=(9,6))  #size of the graph frame

# create a function that creates the graph
def animate(i):
    data = pd.read_csv('sentiment.csv')   # get the dataset into a dataframe
    y1 = data['Trump']
    y2 = data['Biden']
    
    
    if len(y1) <= frame_length: # if the data is less that 10000 rows plot
        plt.cla()
        plt.plot(y1, label='Donald Trump')
        plt.plot(y2, label='Joe Biden')
    else: # if the data is more than 10000, discard the previouse 10000 rows
        plt.cla()
        plt.plot(y1[-frame_length:], label='Donald Trump')
        plt.plot(y2[-frame_length:], label='Joe Biden')
        
    plt.legend(loc='upper left')
    plt.tight_layout()

ani = FuncAnimation(plt.gcf(), animate, interval = 10)
```
